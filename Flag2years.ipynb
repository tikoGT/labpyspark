```python
from pyspark.sql import functions as f

spark = SparkSession.builder.appName("Flag2years").getOrCreate()
```

# Flag para poder indicar, si la información tiene 2 años o más de antiguedad

Carga de CSV

Se descargar el csv, desde un link de mi repositorio en github

```python
!wget https://raw.githubusercontent.com/tikoGT/labpyspark/main/flag2years.csv
```

Lectura del csv

```python
csv_file_path = "/content/flag2years.csv"
dfFlag2years = spark.read.csv(path=csv_file_path, sep=";", header=True)
```

```python
dfFlag2years.withColumn('flag_antiguedad_2', f.when(f.round(f.months_between(f.current_date(),f.col('Fecha_informacion'))/f.lit(12),2) >= 2, 1).otherwise(0)).show(1000)
```
